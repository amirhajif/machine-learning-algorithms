```python
import numpy as np
import matplotlib.pyplot as plt

# تولید داده نمونه
np.random.seed(0)
data = np.random.normal(loc=70, scale=10, size=100)  # داده‌های وزن (میانگین=70، انحراف معیار=10)

# پارامترهای هیستوگرام
num_bins = 10  # تعداد بازه‌ها

# هیستوگرام
counts, bin_edges = np.histogram(data, bins=num_bins, density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# تجسم هیستوگرام
plt.bar(bin_centers, counts, width=bin_edges[1] - bin_edges[0], alpha=0.7, edgecolor='k')
plt.title("Histogram Density Estimation")
plt.xlabel("Weight")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.normal(0, 1, 1000)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].hist(data, bins=50, alpha=0.7, edgecolor='k')
ax[0].set_title("Spiky Histogram (More Bins)")

ax[1].hist(data, bins=10, alpha=0.7, edgecolor='k')
ax[1].set_title("Smooth Histogram (Fewer Bins)")

plt.tight_layout()
plt.show()
```

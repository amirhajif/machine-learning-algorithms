```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_classes=2)

class_1 = X[y == 0]
class_2 = X[y == 1]

mu_1 = np.mean(class_1, axis=0)
mu_2 = np.mean(class_2, axis=0)

S_W = np.cov(class_1.T) + np.cov(class_2.T)

S_B = np.outer(mu_1 - mu_2, mu_1 - mu_2)

w = np.linalg.inv(S_W).dot(mu_1 - mu_2)

plt.scatter(class_1[:, 0], class_1[:, 1], label='Class 1', color='red')
plt.scatter(class_2[:, 0], class_2[:, 1], label='Class 2', color='blue')

x_vals = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_vals = (-w[0] * x_vals - np.dot(w, mu_1 + mu_2) / 2) / w[1]
plt.plot(x_vals, y_vals, label='Decision Boundary', color='black')

plt.legend()
plt.show()
```

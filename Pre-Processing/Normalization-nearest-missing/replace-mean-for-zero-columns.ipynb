```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("pima-indians-diabetes.csv",names=['preg','plas','pres','skin','test','mass','pedi','age','class'])
data
```

## Check Null Data Exist or not

```python
data.isnull()
```

## Count of null items in each Column

```python
data.isnull().sum()
```

## Count of All Null Data in Dataset

```python
data.isnull().sum().sum()
```

## Find a Data with Condition

```python
skines=data[data['skin']==0]
skines
```

```python
mean=data[data['skin']!=0]['skin'].mean()
mean
```

## Replace a value for each column has specific Condition

```python
data.loc[data['skin']==0 ,'skin']=mean
data
```

## Way 2

```python
data.replace({'skin':0},mean,inplace=True)
data
```

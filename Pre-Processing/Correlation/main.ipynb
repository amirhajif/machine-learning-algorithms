```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

this data set contains 146 columns - features

```python
data=pd.read_csv(" .csv")
data.head()
```

```python
data.info()
```

how to find columns with specific Type

```python
object_columns=data.select_dtypes(include=['object'])
object_columns.columns
```

make separate a column data into 2 columns

```python
win_loss_data=data['Win-Loss'].str.split("-",expand=True)
win_loss_data
```

```python
data['Win']=win_loss_data[0]
data['Loss']=win_loss_data[1]
data
```

drop last column

```python
data=data.drop('Win-Loss',axis=1)
data
```

```python
data['Win']=data['Win'].apply(pd.to_numeric)
data['Win']
```

```python
data['Loss']=data['Loss'].apply(pd.to_numeric)
data['Loss']
```

```python
datas_corr=data.corr()
datas_corr
```

```python
datas_corr.loc[['Win']]
```

```python
datas_corr.loc[['Loss']]
```

## Kendal

```python
datas_corr_kendal=data.corr(method='kendall')
datas_corr_kendal
```

```python
datas_corr_kendal['Win']
```

## Plot HeatMap for Corr

```python
plt.Figure(figsize=(20,12))
sns.heatmap(datas_corr,cmap="YlGnBu",annot=True)
```

```python
plt.Figure(figsize=(20,12))
sns.heatmap(datas_corr[datas_corr['Win']>0.7],cmap="YlGnBu",annot=True)
```
